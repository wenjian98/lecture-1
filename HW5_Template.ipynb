# 6.) Using a for loop, make your own train-test split and determine the best max_depth for out-of sample accuracy

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df_select2=df_select.loc[df_select['case_status'].isin(['Certified', 'Denied'])]
```

```python
y=pd.get_dummies(df_select2.case_status)
X=pd.get_dummies(df_select2[["class_of_admission","job_info_experience","ji_live_in_domestic_service","pw_unit_of_pay_9089"]])
```

```python
outputs = []

max_depths = np.arange(1,10)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

for md in max_depths:
  clf = tree.DecisionTreeClassifier(max_depth = md)
  clf.fit(X_train,y_train)

  y_pred = clf.predict(X_test)
  accuracy = accuracy_score(y_test, y_pred)
  outputs.append(accuracy)
best_depths = np.argmax(outputs)
print(best_depths)
```
